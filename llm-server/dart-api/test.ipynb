```python
import pandas as pd

# CSV 파일 불러오기
file_path = '/workspace/008_PseudoLab/CLOVA_Studio_Stock-helper/llm-server/dart-api/samjeon_all.csv'

data = pd.read_csv(file_path)

# 55기 데이터 필터링 (제 55 기)
df_55 = data[data['thstrm_nm'] == '제 55 기']

# 사용할 계정 ID 목록 업데이트
financial_data = df_55[df_55['account_id'].isin([
    'ifrs-full_CurrentAssets', 'ifrs-full_CurrentLiabilities', 'ifrs-full_Liabilities', 'ifrs-full_Equity', 
    'ifrs-full_SharePremium', 'ifrs-full_RetainedEarnings', 'ifrs-full_IssuedCapital', 
    'dart_OperatingIncomeLoss', 'dart_OtherGains', 'dart_OtherLosses', 
    'ifrs-full_ProfitLoss', 'ifrs-full_Revenue', 'ifrs-full_FinanceCosts'
])]

# 계산을 위한 재무 데이터 딕셔너리 생성
financial_dict = financial_data.set_index('account_id')['thstrm_amount'].to_dict()

# 계산을 위해 모든 값을 float로 변환
for key in financial_dict:
    financial_dict[key] = float(financial_dict[key])

# 계산 수행
# 1. 유동비율 = (유동자산 / 유동부채) * 100
current_ratio = (financial_dict['ifrs-full_CurrentAssets'] / financial_dict['ifrs-full_CurrentLiabilities']) * 100

# 2. 부채비율 = (부채총계 / 자본총계) * 100
debt_ratio = (financial_dict['ifrs-full_Liabilities'] / financial_dict['ifrs-full_Equity']) * 100

# 3. 유보율 = (자본잉여금 + 이익잉여금) / 납입자본금 * 100
reserve_ratio = ((financial_dict['ifrs-full_SharePremium'] + financial_dict['ifrs-full_RetainedEarnings']) / 
                 financial_dict['ifrs-full_IssuedCapital']) * 100

# 4. 자본잠식률 = {(자본금 - 자본총계) / 자본금} * 100
capital_impairment_ratio = ((financial_dict['ifrs-full_IssuedCapital'] - financial_dict['ifrs-full_Equity']) / 
                            financial_dict['ifrs-full_IssuedCapital']) * 100

# 5. 경상이익 = 영업이익 + 영업외수익 - 영업외비용
ordinary_income = (financial_dict['dart_OperatingIncomeLoss'] + financial_dict['dart_OtherGains'] - 
                   financial_dict['dart_OtherLosses'])

# 8. 매출액경상이익률 = 경상이익 / 매출액 * 100
ordinary_income_ratio = (ordinary_income / financial_dict['ifrs-full_Revenue']) * 100

# 9. 이자보상배율 = 영업이익 / 이자비용 * 100
interest_coverage_ratio = (financial_dict['dart_OperatingIncomeLoss'] / financial_dict['ifrs-full_FinanceCosts']) * 100

# 10. 자기자본이익률 = 당기순이익 / 자본총액 * 100
roe = (financial_dict['ifrs-full_ProfitLoss'] / financial_dict['ifrs-full_Equity']) * 100

# 결과를 데이터프레임에 저장
results = pd.DataFrame({
    '지표': ['유동비율', '부채비율', '유보율', '자본잠식률', '경상이익', 
           '매출액경상이익률', '이자보상배율', '자기자본이익률'],
    '값': [f"{current_ratio:.2f}%", f"{debt_ratio:.2f}%", f"{reserve_ratio:.2f}%", f"{capital_impairment_ratio:.2f}%", 
          f"{int(ordinary_income)}원", f"{ordinary_income_ratio:.2f}%", f"{interest_coverage_ratio:.2f}%", f"{roe:.2f}%"]
})

results
```

```python
import pandas as pd

# Load the provided CSV file
file_path = '/workspace/008_PseudoLab/CLOVA_Studio_Stock-helper/llm-server/dart-api/samjeon_all.csv'
data = pd.read_csv(file_path, index_col=['Unnamed: 0'])

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Extract necessary columns for calculations
# Select relevant columns based on previous data selection

selected_columns = {
    '자산총계': 'ifrs-full_Assets',
    '유동자산': 'ifrs-full_CurrentAssets',
    '현금및현금성자산': 'ifrs-full_CashAndCashEquivalents',
    '매출채권': 'ifrs-full_TradeAndOtherCurrentReceivables',
    '재고자산': 'ifrs-full_Inventories',
    '자본총계': 'ifrs-full_Equity',
    '부채총계': 'ifrs-full_Liabilities',
    '유동부채': 'ifrs-full_CurrentLiabilities',
    '단기차입금': 'dart_ShortTermBorrowings',
    '장기차입금': 'ifrs-full_NoncurrentBorrowings',
    '사채': 'ifrs-full_IssuedDebtSecurities',
    '영업이익': 'dart_OperatingIncomeLoss',
    '매출총이익': 'ifrs-full_GrossProfit',
    '법인세비용차감전순이익(손실)': 'ifrs-full_ProfitLossBeforeTax',
    '당기순이익(손실)': 'ifrs-full_ProfitLoss',
    '영업수익': 'ifrs-full_Revenue',
    '배당금의 지급': 'ifrs-full_DividendsPaid',
}

# Create a dataframe with the selected columns
df_selected = data[data['account_id'].isin(selected_columns.values())]

# Pivot the table to get years as columns and account names as rows
df_pivot = df_selected.pivot(index='account_id', columns='thstrm_nm', values='thstrm_amount')

# Rename the index based on selected_columns for readability
df_pivot.index = [k for k, v in selected_columns.items() if v in df_pivot.index]

# Display the pivoted dataframe
df_pivot
```

```python
import numpy as np

# Calculate ratios
ratios = {
    '유동비율': (df_pivot.loc['유동자산'] / df_pivot.loc['유동부채']) * 100,
    '자기자본비율': (df_pivot.loc['자본총계'] / df_pivot.loc['자산총계']) * 100,
    '부채비율': (df_pivot.loc['부채총계'] / df_pivot.loc['자본총계']) * 100,
    '매출액 이익율': (df_pivot.loc['매출총이익'] / df_pivot.loc['영업수익']) * 100,
    '총자산 이익율': (df_pivot.loc['영업이익'] / df_pivot.loc['자산총계']) * 100,
    '주당순이익(EPS)': df_pivot.loc['당기순이익(손실)'] / 100,  # Assuming 100 shares for simplicity
    '총자산회전율': (df_pivot.loc['영업수익'] / df_pivot.loc['자산총계']) * 100,
    '배당성향': (df_pivot.loc['배당금의 지급'] / df_pivot.loc['당기순이익(손실)']) * 100
}

# Convert ratios to DataFrame for better readability
ratios_df = pd.DataFrame(ratios, index=['제 55 기'])

ratios_df
```
